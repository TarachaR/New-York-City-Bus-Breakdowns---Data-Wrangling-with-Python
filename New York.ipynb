### Next Steps: Data Cleaning Steps

**Error in values**

- Extract the first integer value (lowest delay time) in the column "How_Long_Delayed"


**Missing Values**

- Impute the missing values in the field "How_Long_Delayed" with the mean value.


**Error in Datatypes**

- Convert "How_Long_Delayed" to int datatype.



**Error in field names**

- Rename the column "Boro" to "Borough".

```
# Lets first start by creating a copy of our dataframe 
# df_clean = df.copy(). We will use this copy as our cleaning copy.
# ---
#
df_clean = df.copy()
df_clean.head()
```

```
# Then extracting the lowest delay time in the column how_long_delayed from the string
# 

df_clean['how_long_delayed'] = df_clean['how_long_delayed'].str.extract('(\d+)')
df_clean['how_long_delayed'].head()
```

We impute the null values in 'how_long_delayed' column with mean of the column. This will take a couple of steps...

```
# We first convert our how_long_delayed to float type to allow for imputation
# ---
# 
df_clean['how_long_delayed']= df_clean['how_long_delayed'].apply(np.float64)
df_clean['how_long_delayed'].dtypes
```

```
# Then later perform our mean imputation 
# ---

df_clean['how_long_delayed'] = df_clean['how_long_delayed'].fillna(df_clean['how_long_delayed'].mean())
df_clean['how_long_delayed'].isnull().any()
```

```
# Then convert back our how_long_delayed column to integer datatype
# ---
# 
df_clean['how_long_delayed'] = df_clean['how_long_delayed'].apply(np.int64)
df_clean['how_long_delayed'].dtypes
```

```
# Then check for nulls in the column
# ---
# 
df_clean['how_long_delayed'].isnull().sum()
```

```
# Rename Boro column to Borough
# ---
# 
df_clean.columns = df_clean.columns.str.replace('boro', 'Borough')
df_clean.columns
```

```
# Lastly we convert all values in our colums to lower case
# for ease of reading
# ---
# 
df_clean.columns = df.columns.str.lower()
df_clean.columns
```

```
# Check the first 5 record the cleaned dataset
# ---
#
df_clean.head()
```


## 5. Solution Implementation

Here we investigate the questions that would help craft our recommendations.

### 5.a) Questions

```
# 1. Which bus companies that had the highest breakdowns?
# ---
#

breakdowns = df_clean.groupby(['bus_company_name', 'breakdown_or_running_late'])[['busbreakdown_id']].count().reset_index()

# Sort to get bus company with highest breakdowns

breakdowns.sort_values(by = 'busbreakdown_id', ascending=0).head(15)
```

```
# 2. What were the top 3 reasons for bus delays?
# ---
#

bus_delay = df_clean.groupby(['reason'])[['how_long_delayed']].count().reset_index()

# Sort to get the most frequent reason

bus_delay.sort_values(by = 'how_long_delayed', ascending = 0).head(3)
```

```
# 3. How many students were in the buses when they broke down? 
# ---
#

number_students = df_clean.groupby(['breakdown_or_running_late'])[['number_of_students_on_the_bus']].sum().reset_index()
number_students.head()
```

```
# 4. Which were most frequent reasons for bus breakdowns?
# ---
#

breakdown_reasons = df_clean.groupby(['breakdown_or_running_late', 'reason'])[['bus_no']].count().reset_index()

# Sort to get most frequent reasons
# ---

breakdown_reasons.sort_values(by='bus_no', ascending=0).head()
```

```
# 5. What were the most frequent reasons for the bus running late? 
# ---
#
reasons = df_clean.groupby(['reason','how_long_delayed']).count()

# Get the records with running late reasons and sort to get most frequent reasons
# ---

reasons.sort_values(by='reason', ascending=0)
```

```
# 6. What was the average delay time for each reason type?
# ---
# 
avg_delay = df_clean.groupby('reason')[['how_long_delayed']].mean().reset_index()

# Get the records with reasons and how long on average a delay took then sorted the records
# ---

avg_delay.sort_values(by='how_long_delayed', ascending=0)
```

### 5.b) Recommendations

From the above analysis, below are our recommendations:

1.   Since most of the delays were caused by Late return from Field Trips, it is recommended that the buses be rotated out to prevent breakdown i.e the buses going out for field trips should be on rotational basis.
2.   Heavy traffic is the most frequent reason for running late and hence either the roads be expanded to ease traffic or the number of vehicles plying that route should be regulated. Alternative routes may also be considered.
3. An investigation into LITTLE RICHIE BUS SERVICE which is the bus company with the most breakdowns should be launched to understand the reason.
